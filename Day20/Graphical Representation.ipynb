```python
import matplotlib.pyplot as plt
```

```python
x = titanic_df.Survived.value_counts()
plt.bar(['Dead','Survived'], x.to_numpy())
plt.show()
plt.close()
```

```python
titanic_df.Survived.value_counts().plot(
    kind = 'barh',
    colormap = 'Set1',
    title = 'Passenger Survivral Rate'
)
```
