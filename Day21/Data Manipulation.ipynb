```python
import pandas as pd
```

```python
titanic_df = pd.read_csv("Titanic_data.csv")
```

```python
titanic_df.shape
```

```python
titanic_df.head()
```

```python
# Dropping column

titanic_df.drop('PassengerId', axis = 1, inplace = True)
titanic_df.head()
```

```python
# Dropping row
```

```python
last_names = [name.split(",")[0] for name in titanic_df.Name]
titanic_df['First Name'] = last_names
```

```python
first_names = [name.split(",")[1] for name in titanic_df.Name]
titanic_df['First Name'] = first_names
```

```python
titanic_df.head()
```

```python
first_names = [name.split(",")[1][0:5] for name in titanic_df.Name]
first_names
```

```python
Salutation = [name.split(",")[1].split(".")[0] for name in titanic_df.Name]
titanic_df['Salutation'] = Salutation
```

```python
titanic_df.head()
```

```python
# inserting in a row

titanic_df.loc[0,'Cabin'] = 'C23'
```
